```python
import pandas as pd
df=pd.read_csv('/kaggle/input/titanic-decision-tree/titanic.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labeled=LabelEncoder()
df['Sex']=labeled.fit_transform(df['Sex'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df['Age']=df['Age'].fillna(df['Age'].median())
x=df[['Pclass','Sex','Age','Fare']]

y=df['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=DecisionTreeClassifier(max_depth=5,min_samples_split=10,random_state=42)
model.fit(x_train,y_train)
pred=model.predict(x_test)

print("Accuracy:", accuracy_score(y_test,pred))
score=cross_val_score(model,x,y,cv=5)
print('score:',score.mean())
user1=float(input('enter user pclas:'))
user2=input('enter user sex:')
user3=float(input('enter user age:'))
user4=float(input('enter user fare:'))
user2_enc=labeled.transform([user2])[0]
user_features=[[user1,user2_enc,user3,user4]]
user_features = pd.DataFrame([[user1, user2_enc, user3, user4]],
                             columns=['Pclass','Sex','Age','Fare'])
print('Survived' if pred[0]==1 else 'Not Survived')
```
